```python
#importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as pl
import matplotlib.pyplot as plt
```

```python
#Loading the dataset

data=pd.read_csv("Unemployment in India.csv")
data
```

```python
#checking for the null values

data.isna().sum()
```

```python
#Removing the null values

data = data.dropna()
```

```python
#checking for the null values now

data.isna().sum()
```

```python
data.columns = data.columns.str.strip()

# Print cleaned column names to verify
print(data.columns)
```

```python
# Checking for duplicate entries in the dataset

data.duplicated().sum()
```

```python
# printing the columns of the dataset

data.columns
```

```python
# Counting the values of Area

data['Area'].value_counts()
```

```python
#Distribution of urban and rural areas

area_counts = data['Area'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(area_counts, labels=area_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Urban and Rural Areas')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Counting the values of regions

data['Region'].value_counts()
```

```python
# Count Plot Representing the  regions

sns.countplot(x = data['Region'], palette = "bright")
plt.xticks(rotation ='vertical')
plt.show()
```

```python
columns_of_interest = ['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']
data_filtered = data[columns_of_interest]
```

```python
# Checking for missing values
print(data_filtered.isnull().sum())

# Optionally, filling or droping missing values
data_filtered = data_filtered.dropna()
```

```python
correlation_matrix = data_filtered.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Unemployment Metrics')
plt.show()
```

```python
# Box Plot of Estimated Unemployment Rate

plt.figure(figsize=(24, 7))
sns.boxplot(x = 'Region', y= 'Estimated Unemployment Rate (%)', data = data)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# Box Plot of Estimated employed

plt.figure(figsize=(24, 7))
sns.boxplot(x = 'Region', y= 'Estimated Employed', data = data)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# Box Plot of Estimated Labour Participation Rate 

plt.figure(figsize=(24, 7))
sns.boxplot(x = 'Region', y= 'Estimated Labour Participation Rate (%)', data = data)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
frequency_counts = data['Date'].value_counts()
frequency_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(frequency_counts, labels=frequency_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Date')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
